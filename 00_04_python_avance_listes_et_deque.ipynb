```python
import time
import pandas as pd
```

```python
def duree(debut):
    return time.time() - debut
```

# Dur√©e d'insertion en queue de liste

```python
conclusion = u"L'insertion en queue de liste est en temps constant"
liste      = list()
durees     = {}
for i in range(100000):
    debut = time.time()
    liste.append(i)
    fin   = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

# Duree d'insertion en t√™te de liste

```python
conclusion         = u"L'insertion en t√™te de liste est en temps lin√©aire"
liste              = list()
position_insertion = 0
durees             = {}
for i in range(100000):
    debut = time.time()
    liste.insert(position_insertion, i)
    fin = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```

# La classe "dequeue"

```python
from collections import deque
conclusion         = u"""
Deque insert en t√™te de liste en temps constant
et en √©tant  un ordre de magnitude plus rapide.
"""
double_queue       = deque()
position_insertion = 0
durees             = {}
for i in range(100000):
    debut = time.time()
    double_queue.appendleft(i)
    fin = time.time()
    durees[i] = fin - debut
    
df = pd.DataFrame.from_dict(durees, orient='index')    
df.plot(title=conclusion)
```
