```python
def get_row():
    import random
    branch  = ["cib", "wealth", "retail"]
    score   = [ "blank", "nsc", "high", "low"]
    actif   = random.randint(0, 1000000)
    passif  = random.randint(0,1000000)
    CLIENT_ID  = random.randint(0,1000000)
    a_p     = actif + passif
    r = {"branch" : random.choice(branch),
        "score":random.choice(score),
         "actif" :  actif ,
         "passif" : passif ,
         "actif + passif" : a_p,
        "CLIENT_ID" : CLIENT_ID}
    return r
```

```python
import pandas as pd
df = pd.DataFrame([get_row() for i in range(2000)])
```

```python
df
```

```python
import numpy as np
```

```python
pt = pd.pivot_table(df[["branch", "score", "actif", "passif", "actif + passif"]], 
               index=["branch", "score"], 
               aggfunc=[np.min, np.max, np.mean, np.std])
pt
```

```python
%matplotlib inline
ax = pt.T.plot.box(rot=45, legend=False)
```

```python
%matplotlib inline
ax = pt.T.plot.box(rot=45, legend=False)
a = list([b.get_text()[b.get_text().find(",")+1:-1] for b in ax.get_xticklabels()])
ax.set_xticklabels(a)
```
